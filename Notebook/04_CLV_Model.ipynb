```python
import os
import pandas as pd
import joblib
from sklearn.linear_model import LinearRegression

# Load RFM data
rfm = pd.read_csv(r"E:\Personal Project\customer-intelligence-system\Data\rfm_data.csv")

X = rfm[['Recency','Frequency']]
y = rfm['Monetary']

# Train model
model = LinearRegression()
model.fit(X,y)

# Ensure 'models' folder exists
model_folder = r"E:\Personal Project\customer-intelligence-system\models"
os.makedirs(model_folder, exist_ok=True)

# Save CLV model
joblib.dump(model, os.path.join(model_folder, "clv_model.pkl"))

print("CLV model saved successfully ✅")
```
